```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data_hechos = pd.read_excel('homicidios.xlsx',sheet_name='HECHOS')
data_victimas = pd.read_excel('homicidios.xlsx',sheet_name='VICTIMAS')
```

```python
# Se cre una funcion para detectar los nulos en los distintos campos del dataframe y si hay vacios 
# contarlos por nulos
def count_missing_and_empty_values(df):
    # Replace empty strings with NaN so they can be counted
    df.replace('', np.nan, inplace=True)
    
    # Calculate the count of missing and empty values for each column
    missing_counts = df.isna().sum()
    
    # Create a DataFrame to display the results
    missing_info = pd.DataFrame({'Column Name': missing_counts.index, 'Missing/Empty Values': missing_counts.values})

    return missing_info
```

```python
count_missing_and_empty_values(data_hechos)
```
